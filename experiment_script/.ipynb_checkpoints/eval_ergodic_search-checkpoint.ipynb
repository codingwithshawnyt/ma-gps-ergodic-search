# MA-GPS Ergodic Search Evaluation
Run this notebook from the `experiment_script/` directory.

```python
import argparse
import os
import numpy as np
import torch
import gymnasium as gym
from torch.distributions import Independent, Normal
from scipy.stats import multivariate_normal as mvn

from MAGPS.data import Batch
from MAGPS.policy.gym_marl_policy.ippo import IPPOPolicy
from MAGPS.utils.net.common import ActorCritic, Net
from MAGPS.utils.net.continuous import ActorProb, Critic

# ---- Configure ----
DEVICE = "cpu"
TASK = "ergodic_search-v0"
EPOCH_TO_LOAD = 15

args = argparse.Namespace(
    task=TASK,
    seed=0,
    buffer_size=40000,
    lr=1e-4,
    gamma=0.99,
    epoch=5,
    total_episodes=30,
    step_per_epoch=40000,
    episode_per_collect=16,
    batch_size=2048,
    critic_net=[512, 512, 512],
    actor_net=[512, 512, 512],
    training_num=64,
    test_num=100,
    logdir="log",
    behavior_loss_weight=0.1,
    behavior_loss_weight_decay=1,
    regularization=False,
    no_gd_regularization=False,
    device=DEVICE,
    repeat_per_collect=10,
    vf_coef=0.25,
    ent_coef=0.005,
    eps_clip=0.2,
    max_grad_norm=0.5,
    gae_lambda=0.99,
    rew_norm=1,
    dual_clip=None,
    value_clip=1,
    norm_adv=1,
    recompute_adv=0,
    kwargs="{}",
)

env = gym.make(args.task)
args.max_action = env.action_space.high[0]
args.state_shape = env.observation_space.shape or env.observation_space.n
args.action_shape = env.action_space.shape or env.action_space.n
args.action_shape_per_player = env.nu

actor_net = Net(args.state_shape, hidden_sizes=args.actor_net, device=args.device)
actor_list = [
    ActorProb(
        actor_net, args.action_shape_per_player, max_action=args.max_action, device=args.device
    ).to(args.device) for i in range(env.num_players)
]
critic_net = Net(args.state_shape, hidden_sizes=args.critic_net, device=args.device)
critic_list = [
    Critic(critic_net, device=args.device).to(args.device) for i in range(env.num_players)
]

def dist(*logits):
    return Independent(Normal(*logits), 1)

policy = IPPOPolicy(
    actor_list,
    critic_list,
    [torch.optim.Adam(ActorCritic(actor_list[i], critic_list[i]).parameters(), lr=args.lr) for i in range(env.num_players)],
    env.num_players,
    env.nu,
    dist,
    discount_factor=args.gamma,
    max_grad_norm=args.max_grad_norm,
    eps_clip=args.eps_clip,
    vf_coef=args.vf_coef,
    ent_coef=args.ent_coef,
    reward_normalization=args.rew_norm,
    advantage_normalization=args.norm_adv,
    recompute_advantage=args.recompute_adv,
    dual_clip=args.dual_clip,
    value_clip=args.value_clip,
    gae_lambda=args.gae_lambda,
    action_space=env.action_space,
    device=args.device,
    deterministic_eval=True,
    pure_policy_regulation=args.regularization,
    no_gd_regularization=args.no_gd_regularization,
    env=env,
    batch_size=args.batch_size,
)

log_path = os.path.join(
    args.logdir, args.task,
    'ippo_training_num_{}_buffer_size_{}_c_{}_{}_a_{}_{}_gamma_{}_behavior_loss_{}_{}_L2_reg_{}'.format(
        args.training_num, args.buffer_size,
        args.critic_net[0], len(args.critic_net),
        args.actor_net[0], len(args.actor_net),
        args.gamma, args.behavior_loss_weight,
        args.behavior_loss_weight_decay, args.regularization
    )
)
log_path = log_path + '/lr_{}_batch_{}_step_per_epoch_{}_kwargs_{}_seed_{}'.format(
    args.lr, args.batch_size, args.step_per_epoch, args.kwargs, args.seed
)

policy_file = os.path.join(log_path, f"epoch_id_{EPOCH_TO_LOAD}", "policy.pth")
print(f"Loading policy from: {policy_file}")
print(f"File exists: {os.path.exists(policy_file)}")
policy.load_state_dict(torch.load(policy_file, map_location=DEVICE))
policy.eval()
print("Policy loaded successfully!")
```

```python
# ---- Simulate trajectory ----

def find_a(state):
    tmp_obs = np.array(state).reshape(1, -1)
    tmp_batch = Batch(obs=tmp_obs, info=Batch())
    with torch.no_grad():
        tmp = policy(tmp_batch, model="actor_old").act
    act = policy.map_action(tmp).cpu().detach().numpy().flatten()
    return act

horizon = 200
n = env.observation_space.shape[0]
m = env.action_space.shape[0]
x_traj = np.zeros((n, horizon + 1))
u_traj = np.zeros((m, horizon))
metrics = []

# Random initial positions
np.random.seed(42)
x0 = np.random.uniform(low=0.2, high=0.8, size=n)
x_traj[:, 0] = x0

env.reset(options={"initial_state": x0})

for t in range(horizon):
    u_traj[:, t] = find_a(x_traj[:, t])
    obs, reward, terminated, truncated, info = env.step(u_traj[:, t])
    x_traj[:, t + 1] = obs
    metrics.append(info['ergodic_metric'])
    if terminated:
        print(f"Agent left bounds at t={t}")
        break

actual_horizon = len(metrics)
print(f"Simulated {actual_horizon} steps")
print(f"Final ergodic metric: {metrics[-1]:.6f}")
for i in range(env.num_players):
    print(f"  Agent {i+1} final pos: ({x_traj[i*2, actual_horizon]:.3f}, {x_traj[i*2+1, actual_horizon]:.3f})")
```

```python
# ---- Static plot: trajectories on target distribution ----
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Build target PDF heatmap
grid_res = 200
gx = np.linspace(0, 1, grid_res)
gy = np.linspace(0, 1, grid_res)
GX, GY = np.meshgrid(gx, gy)
grid_pts = np.column_stack([GX.ravel(), GY.ravel()])

mean1, cov1, w1 = np.array([0.35, 0.38]), np.array([[0.01, 0.004], [0.004, 0.01]]), 0.5
mean2, cov2, w2 = np.array([0.68, 0.25]), np.array([[0.005, -0.003], [-0.003, 0.005]]), 0.2
mean3, cov3, w3 = np.array([0.56, 0.64]), np.array([[0.008, 0.0], [0.0, 0.004]]), 0.3

pdf_vals = (w1 * mvn.pdf(grid_pts, mean1, cov1) +
            w2 * mvn.pdf(grid_pts, mean2, cov2) +
            w3 * mvn.pdf(grid_pts, mean3, cov3))
PDF = pdf_vals.reshape(grid_res, grid_res)

colors = ["#E63946", "#457B9D", "#2A9D8F"]
labels = [f"Agent {i+1}" for i in range(env.num_players)]

fig, ax = plt.subplots(figsize=(8, 8))

# Target distribution
ax.contourf(GX, GY, PDF, levels=30, cmap="YlOrRd", alpha=0.4)
ax.contour(GX, GY, PDF, levels=10, colors="gray", linewidths=0.5, alpha=0.5)

# Gaussian peaks
for m, w, label in [(mean1, w1, "Peak 1 (w=0.5)"), (mean2, w2, "Peak 2 (w=0.2)"), (mean3, w3, "Peak 3 (w=0.3)")]:
    ax.plot(m[0], m[1], 'k+', markersize=15, markeredgewidth=2)
    ax.annotate(label, (m[0], m[1]), textcoords="offset points", xytext=(8, 8), fontsize=8, color="gray")

# Agent trajectories
T = actual_horizon + 1
for i in range(env.num_players):
    xi = x_traj[i*2, :T]
    yi = x_traj[i*2+1, :T]
    ax.plot(xi, yi, color=colors[i], linewidth=1.5, alpha=0.7, label=labels[i])
    ax.plot(xi[0], yi[0], 'o', color=colors[i], markersize=12, markeredgecolor='white', markeredgewidth=2, zorder=5)
    ax.plot(xi[-1], yi[-1], 's', color=colors[i], markersize=10, markeredgecolor='white', markeredgewidth=2, zorder=5)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend(loc='upper right')
ax.set_title(f'MA-GPS Ergodic Search (epoch {EPOCH_TO_LOAD})\nFinal metric: {metrics[-1]:.4f}')
plt.tight_layout()
plt.show()
```

```python
# ---- Ergodic metric over time ----
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(metrics, linewidth=2, color='#E63946')
ax.set_xlabel('Timestep')
ax.set_ylabel('Ergodic Metric')
ax.set_title('Ergodic Metric Over Episode')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ---- Animation ----
import matplotlib.animation as animation
from IPython.display import HTML

TRAIL_LEN = 50
T = actual_horizon + 1

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Left: agent trajectories on target distribution
ax1.contourf(GX, GY, PDF, levels=30, cmap="YlOrRd", alpha=0.4)
ax1.contour(GX, GY, PDF, levels=10, colors="gray", linewidths=0.5, alpha=0.5)
for m in [mean1, mean2, mean3]:
    ax1.plot(m[0], m[1], 'k+', markersize=12, markeredgewidth=2)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_aspect('equal')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

trails = []
dots = []
for i in range(env.num_players):
    trail, = ax1.plot([], [], color=colors[i], linewidth=2, alpha=0.5, label=labels[i])
    dot, = ax1.plot([], [], 'o', color=colors[i], markersize=12, markeredgecolor='white', markeredgewidth=2, zorder=5)
    trails.append(trail)
    dots.append(dot)
ax1.legend(loc='upper right', fontsize=8)
title1 = ax1.set_title('Ergodic Search  t=0')

# Right: ergodic metric over time
ax2.set_xlim(0, actual_horizon)
ax2.set_ylim(0, max(metrics) * 1.1)
ax2.set_xlabel('Timestep')
ax2.set_ylabel('Ergodic Metric')
ax2.grid(True, alpha=0.3)
metric_line, = ax2.plot([], [], color='#E63946', linewidth=2)
title2 = ax2.set_title('Ergodic Metric')

def init():
    for trail in trails:
        trail.set_data([], [])
    for dot in dots:
        dot.set_data([], [])
    metric_line.set_data([], [])
    return trails + dots + [metric_line]

def animate(frame):
    t = frame
    t0 = max(0, t - TRAIL_LEN)
    for i in range(env.num_players):
        xi = i * 2
        yi = i * 2 + 1
        trails[i].set_data(x_traj[xi, t0:t+1], x_traj[yi, t0:t+1])
        dots[i].set_data([x_traj[xi, t]], [x_traj[yi, t]])
    metric_line.set_data(range(t), metrics[:t])
    title1.set_text(f'Ergodic Search  t={t}/{actual_horizon}')
    return trails + dots + [metric_line, title1]

anim = animation.FuncAnimation(fig, animate, init_func=init,
                                frames=actual_horizon, interval=40, blit=True)

anim.save("ergodic_search.mp4", writer="ffmpeg", fps=25, dpi=150)
print("Saved ergodic_search.mp4")

HTML(anim.to_jshtml())
```

```python
# ---- Compare different checkpoints ----
# Change EPOCH_TO_LOAD above and re-run, or run this cell to compare multiple epochs

epochs_to_compare = [15, 30, 60, 90]
fig, axes = plt.subplots(1, len(epochs_to_compare), figsize=(5 * len(epochs_to_compare), 5))

for idx, ep in enumerate(epochs_to_compare):
    ax = axes[idx]
    pf = os.path.join(log_path, f"epoch_id_{ep}", "policy.pth")
    if not os.path.exists(pf):
        ax.set_title(f'Epoch {ep}\n(not found)')
        continue
    
    policy.load_state_dict(torch.load(pf, map_location=DEVICE))
    policy.eval()
    
    # Simulate
    xt = np.zeros((n, horizon + 1))
    xt[:, 0] = x0.copy()
    env.reset(options={"initial_state": x0.copy()})
    ep_metrics = []
    
    for t in range(horizon):
        a = find_a(xt[:, t])
        obs, _, term, _, info = env.step(a)
        xt[:, t+1] = obs
        ep_metrics.append(info['ergodic_metric'])
        if term:
            break
    
    T = len(ep_metrics) + 1
    
    ax.contourf(GX, GY, PDF, levels=20, cmap="YlOrRd", alpha=0.3)
    for i in range(env.num_players):
        ax.plot(xt[i*2, :T], xt[i*2+1, :T], color=colors[i], linewidth=1.5, alpha=0.7)
        ax.plot(xt[i*2, 0], xt[i*2+1, 0], 'o', color=colors[i], markersize=8, markeredgecolor='white')
    for m in [mean1, mean2, mean3]:
        ax.plot(m[0], m[1], 'k+', markersize=10, markeredgewidth=2)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.set_title(f'Epoch {ep}\nMetric: {ep_metrics[-1]:.4f}' if ep_metrics else f'Epoch {ep}')

plt.suptitle('MA-GPS Ergodic Search: Training Progress', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
